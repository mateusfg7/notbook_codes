```
import matplotlib.pyplot as plt

def exibirMateria(materia, legenda, cor):
  # plt.plot(eixo, materia, color='k', linestyle=':')
  # plt.scatter(eixo, materia, label=legenda, color=cor)

  eixo_x = ['1ºB '+legenda[:2],'2ºB '+legenda[:2],'3ºB '+legenda[:2]]
  plt.bar(eixo_x, materia, label=legenda, color=cor)


artes = [25, 17, 18]
filosofia = [18,24,17]
geografia = [16,24,19]
historia = [20,23,16]
ingles = [18,17,15]
portugues = [15,17,19]
sociologia = [12,15,24]

biologia = [18, 16, 19]
edfisica = [22, 20, 22]

fisica = [22,21,25]
matematica = [21,15,23]
quimica = [6,20,20]


exibirMateria(artes, 'Artes', '#ff4a4a')
exibirMateria(filosofia, 'Filosofia', '#ff4a7d')
exibirMateria(geografia, 'Geografia', '#ff4aa4')
exibirMateria(historia, 'História', '#ff4ac6')
exibirMateria(ingles, 'Inglês', '#ff4aea')
exibirMateria(portugues, 'Português', '#d84aff')
exibirMateria(sociologia, 'Sociologia', '#ab4aff')

exibirMateria(fisica, 'Física', '#4a53ff')
exibirMateria(matematica, 'Matématica', '#4a9eff')
exibirMateria(quimica, 'Química', '#4adbff')

exibirMateria(biologia, 'Biologia', '#4affbd')
exibirMateria(edfisica, 'Educação Física', '#4aff80')

plt.rcParams['figure.figsize'] = (20,7)
plt.legend()
plt.show()
```
